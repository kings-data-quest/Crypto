```python
engine = create_engine('mysql://root:kingsley@127.0.0.1:3306/Binance_futures_data')
```

```python
def qry(symbol, lookback = 1090):
    now = dt.datetime.utcnow()
    before = str(now - dt.timedelta(weeks= lookback))
    qry_str = f"SELECT * FROM Binance_futures_data.{symbol} WHERE Date >= '{before}'"
    df = pd.read_sql(qry_str, engine)
    df = df.set_index('Date')
    return df
```
